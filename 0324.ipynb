```python
### 程式實例ch3_24.py:從地球到月球約是384400公里，假設或見的速度是一馬赫，設計一個程式計算需要多少天、多少小時才可抵達月球。這個程式省略分鐘數。
```

```python
# ch3_24.py
dist = 384400                 #地球到月亮距離
speed = 1225                  #馬赫速度每小時1225公里
total_hours = dist // speed   #計算小時數
days = total_hours // 24      #商 = 計算天數
hours = total_hours % 24      #餘數 = 計算小時數
print("總共需要天數")
print(days)
print("小時數")
print(hours)
```

```python
# ch5_4.py
print("奇數偶數判斷")
num = input("請輸入任意數值: ")
rem = int(num) % 2
if (rem == 0):
    print("%d 是偶數" % int(num))
else:
     print("%d 是奇數" % int(num))
```

```python
# ch5_4.py
print("3 的倍數判斷")
num = input("請輸入任意數值: ")
rem = int(num) % 3
if (rem == 0):
    print("%d 是3的倍數" % int(num))
else:
     print("%d 不是3的倍數" % int(num))
```
