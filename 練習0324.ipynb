```python
### 城市實例ch3_24.py:從地球到月球約是384400公里，假設火箭的速度是一馬赫，設計一個程式計算需要多少天，多少小時才可抵達月球。這個程式省略分鐘數
```

```python
# ch3_24.py
dist = 384400
speed = 1225
total_hours = dist // speed
days = total_hours // 24
hours = total_hours % 24
print("總共需要天數")
print(days)
print("小時數")
print(hours)
```

```python
# ch5_4.py
print("奇數偶數判斷")
num =input("請輸入任意整值: ")
rem = int(num) % 2
if (rem == 0):
    print("%d 是偶數" % int(num))
else:
    print("%d 是奇數" % int(num))
```
