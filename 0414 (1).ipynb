```python
score = [80, 90, 85, 75, 70, 65, 88, 98, 100]
index = score.index(75)
score[index] = 78
print("最高分:", max(score))
print("最低分:", min(score))
print("總分:", sum(score))
score.append(50)
score.append(99)
print("加入後:", score)
score.sort()
print("小到大:", score)
score.sort(reverse=True)
print("大到小", score)
```
