```python
score = [80, 90, 85, 75, 70, 65, 88, 98, 100]
score[3] = 78
print("最高分 =", max(score))
print("最低分 =", min(score))
print("總分 =", sum(score))
score.append(50)
score.append(99)
print("新增後的串列 =", score)
score.sort()
print("由小到大排序 =", score)
score.sort(reverse=True)
print("由大到小排序 =", score)
```
