```python
# ch5_5.py
print("計算最終成績")
score = input("請輸入分數: ")
sc = int(score)
if (sc >= 90):
    print(" A")
elif (sc >= 80):
    print(" B")
elif (sc >= 70):    
    print(" C")
elif (sc >= 60): 
    print(" D")
else:
    print(" F")
```

```python
year_input = input("請輸入出生西元年: ")
year = int(year_input)
# 計算餘數
rem = (year - 1900) % 12
# 判斷生效
if rem == 0:
    print("生肖屬:鼠")
if rem == 1:
    print("生肖屬:牛")
if rem == 2:
    print("生肖屬:虎")
if rem == 3:
    print("生肖屬:兔")
if rem == 4:
    print("生肖屬:龍")
if rem == 5:
    print("生肖屬:蛇")
if rem == 6:
    print("生肖屬:馬")
if rem == 7:
    print("生肖屬:羊")
if rem == 8:
    print("生肖屬:猴")
if rem == 9:
    print("生肖屬:雞")
if rem == 10:
    print("生肖屬:狗")
else:
    print("生肖屬:豬")

                            
                            
```

```python
score = [100, 80, 90, 70, 85, 88, 77, 60, 50, 97]
#取得第二位(索引1)到第七位(索引6)的同學
#語法是 [1:7],因為7不包含,所以會抓到索引1, 2, 3, 4, 5, 6
target_scores = score[1:7]
print("第二位到第七位同學的最高分 =",max(target_scores))
print("第二位到第七位同學的最低分 =",min(target_scores))
```
